##Perceptron przykÅ‚ad

```python
import numpy as np

x = np.array(
    [[0,0],
    [0,1],
    [1,0],
    [1,1]])

d = np.array([0,0,0,1])

w = np.random.random(3)
```

```python
for i in range( len(x) ):
  xx = x[i]
  dd = d[i]

  s = xx[0]*w[1] + xx[1]*w[2] + w[0]*(-1)
  if s >= 0:
    y = 1
  else:
    y = 0

  print("x1: ", xx[0], " x2: ", xx[1], " d: ", dd, " y: ", y)
```

```python
mi = 0.1
for a in range(1):
  for i in range( len(x) ):
    xx = x[i]
    dd = d[i]

    s = xx[0]*w[1] + xx[1]*w[2] + w[0]*(-1)
    if s >= 0:
      y = 1
    else:
      y = 0

    w[0] = w[0] + mi*(dd-y)*(-1)
    w[1] = w[1] + mi*(dd-y)*xx[0]
    w[2] = w[2] + mi*(dd-y)*xx[1]
```

```python
import matplotlib.pyplot as plt
xx = np.arange(-1,3)
yy = -(w[1]/w[2]) * xx + (w[0]/w[2])
plt.plot(xx,yy, 'r-')

for i in range( len(x) ):
  if d[i] == 0:
    plt.plot(x[i, 0], x[i, 1], 'ko')
  else:
    plt.plot(x[i, 0], x[i, 1], 'm*')
```

Nauka preceptronu tylko do momentu nauczenia siÄ™ punktÃ³w

```python
w = np.random.random(3)
import matplotlib.pyplot as plt

mi = 0.003
max_iterations = 100

errors_per_iteration = []

for a in range(max_iterations):
    errors = 0

    for i in range(len(x)):
        xx = x[i]
        dd = d[i]

        s = xx[0] * w[1] + xx[1] * w[2] + w[0] * (-1)
        y = 1 if s >= 0 else 0

        if y != dd:
            errors += 1
            w[0] = w[0] + mi * (dd - y) * (-1)
            w[1] = w[1] + mi * (dd - y) * xx[0]
            w[2] = w[2] + mi * (dd - y) * xx[1]

    errors_per_iteration.append(errors)

    if errors == 0:
        print(f"Perceptron nauczony po {a+1} iteracjach.")
        break
else:
    print("OsiÄ…gniÄ™to maksymalnÄ… liczbÄ™ iteracji bez peÅ‚nego nauczenia.")


plt.plot(range(1, len(errors_per_iteration) + 1), errors_per_iteration, marker='o')
plt.xlabel("Iteracja")
plt.ylabel("Liczba bÅ‚Ä™dÃ³w")
plt.title("Liczba bÅ‚Ä™dÃ³w podczas uczenia perceptronu")
plt.grid()
plt.show()
```
